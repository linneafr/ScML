```python
import csv
import pandas as pd


        
df_air_quality = pd.read_csv('grensasv_iceland-air-quality.csv', error_bad_lines=False)
df_air_quality['date'].str.strip()
df_air_quality['pm25'].str.strip()
df_air_quality['pm10'].str.strip()
df_air_quality['no2'].str.strip()
df_air_quality['co'].str.strip()
    
df_air_quality.head()
```

```python
print(df_air_quality.columns)
df_air_quality.date = df_air_quality.date.apply(pd.to_datetime)
print(df_air_quality.info())
df_air_quality.head()
```

```python
from datetime import datetime

def timestamp_2_time(x):
    dt_obj = datetime.strptime(str(x), '%Y-%m-%d')
    dt_obj = dt_obj.timestamp() * 1000
    return int(dt_obj)

df_air_quality.date
```

```python
df_air_quality.date = df_air_quality.date.apply(timestamp_2_time)
df_air_quality.head()
```

```python
df_weather.datetime = df_weather.datetime.apply(timestamp_2_time)
df_weather.head()
```
